```
from six import BytesIO
from PIL import Image
import numpy as np
import glob
import os

import matplotlib
import matplotlib.pyplot as plt
```

```
def load_images_into_numpy_array(path):
  img_data = open(path,'rb').read()
  image = Image.open(BytesIO(img_data))
  (im_width,im_height) = image.size
  return np.array(image.getdata()).reshape((im_width,im_height,3)).astype(np.uint8)
```

```
elephant_image_path = "/content/image/elephant is africa/*"
elephant_image_np = []
for iname in glob.glob(elephant_image_path):
  elephant_image_np.append(load_images_into_numpy_array(iname))
```

```
plt.rcParams['figure.figsize'] =[14,7]
for idx, elephant_image_np in enumerate(elephant_image_np):
  plt.subplot(2,3,idx+1)
  plt.imshow(elephant_image_np)
plt.show()
```

```
kangaroo_image_path = "/content/images/kangaroo in australia/*"
kangaroo_image_np = []
for iname in glob.glob(kangaroo_image_path):
  kangaroo_image_np.append(load_images_into_numpy_array(iname))
```

```
plt.rcParams['figure.figsize'] =[14,7]

for idx, kangaroo_image_np in enumerate(kangaroo_image_np):
  plt.subplot(2,3,idx+1)
  plt.imshow(kangaroo_image_np)

plt.show()
```
